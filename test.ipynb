```python
import os
```

```python
base_path = "C:\\Users\\DELL\\Documents\\S9\\Projet_Pic\\dead-bird-detection\\augmented_dataset"

def analyze_labels(split_path):
    label_dir = os.path.join(split_path, "labels")

    total_files = 0
    empty_files = 0
    files_with_class_0 = 0

    for label_file in os.listdir(label_dir):
        if not label_file.endswith(".txt"):
            continue

        total_files += 1
        file_path = os.path.join(label_dir, label_file)

        with open(file_path, "r") as f:
            lines = [line.strip() for line in f.readlines() if line.strip()]

        # Fichier vide ou sans annotations
        if len(lines) == 0:
            empty_files += 1
            continue

        # Vérifier présence de la classe 0
        has_class_0 = any(line.split()[0] == "0" for line in lines)
        if has_class_0:
            files_with_class_0 += 1

    return total_files, empty_files, files_with_class_0


for split in ["train", "valid", "test"]:
    split_path = os.path.join(base_path, split)
    total, empty, class_0 = analyze_labels(split_path)

    print(f"\n {split.upper()}")
    print(f"Total fichiers labels : {total}")
    print(f"Fichiers vides         : {empty}")
    print(f"Fichiers avec classe 0 : {class_0}")
```
