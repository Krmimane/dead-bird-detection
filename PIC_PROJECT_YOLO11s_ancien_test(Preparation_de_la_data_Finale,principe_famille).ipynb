```python
import os
import re
import random
import shutil
from pathlib import Path

# =========================
# CONFIG
# =========================
INPUT_ROOT  = "/content/drive/MyDrive/PIC_2_Corrtldbbox_tiled_512_augmented"
OUTPUT_ROOT = "/content/drive/MyDrive/PIC_FINAL_512_NOLEAK2"

TRAIN_P = 0.65
VAL_P   = 0.18
TEST_P  = 0.17

BG_RATIO_TRAIN = 1   # 1:1
BG_RATIO_VAL   = 2   # 2:1
BG_RATIO_TEST  = 2   # 2:1

SEED = 42
IMAGE_EXTS = (".jpg", ".jpeg", ".png")

SPLITS_IN  = ["train", "valid", "test"]
SPLITS_OUT = ["train", "valid", "test"]

random.seed(SEED)

AUG_RE = re.compile(r"(.*)_aug\d+$")   # dÃ©tecte base_augN


# =========================
# HELPERS
# =========================
def is_positive_label(lbl_path: Path) -> bool:
    return lbl_path.exists() and lbl_path.stat().st_size > 0


def is_augmented(stem: str) -> bool:
    return AUG_RE.match(stem) is not None


def base_id(stem: str) -> str:
    """Regroupe original + augmentations sous un mÃªme identifiant."""
    m = AUG_RE.match(stem)
    return m.group(1) if m else stem


def ensure_dirs(root: Path):
    for sp in SPLITS_OUT:
        (root / sp / "images").mkdir(parents=True, exist_ok=True)
        (root / sp / "labels").mkdir(parents=True, exist_ok=True)


def collect_all_pairs_grouped(input_root: Path):
    """
    Retourne:
      pos_groups[base] = list[(img_path,lbl_path,is_aug)]
      bg_groups[base]  = list[(img_path,lbl_path,is_aug)]
    """
    pos_groups = {}
    bg_groups  = {}

    for sp in SPLITS_IN:
        img_dir = input_root / sp / "images"
        lbl_dir = input_root / sp / "labels"
        if not img_dir.exists():
            continue

        for img_name in os.listdir(img_dir):
            if not img_name.lower().endswith(IMAGE_EXTS):
                continue

            img_path = img_dir / img_name
            stem = img_path.stem
            bid = base_id(stem)

            lbl_path = lbl_dir / (stem + ".txt")
            aug_flag = is_augmented(stem)

            if is_positive_label(lbl_path):
                pos_groups.setdefault(bid, []).append((img_path, lbl_path, aug_flag))
            else:
                bg_groups.setdefault(bid, []).append((img_path, lbl_path, aug_flag))

    return pos_groups, bg_groups


def split_bases(bases, p_train, p_val, p_test):
    bases = list(bases)
    random.shuffle(bases)

    n = len(bases)
    n_train = int(n * p_train)
    n_val   = int(n * p_val)

    train = bases[:n_train]
    val   = bases[n_train:n_train+n_val]
    test  = bases[n_train+n_val:]

    return set(train), set(val), set(test)


def pick_background_bases(bg_bases, n_needed):
    bg_bases = list(bg_bases)

    if n_needed <= 0:
        return set()
    if n_needed >= len(bg_bases):
        return set(bg_bases)

    return set(random.sample(bg_bases, n_needed))


def safe_copy(
    img_path: Path,
    lbl_path: Path,
    out_img_dir: Path,
    out_lbl_dir: Path,
    used_names: set
):
    base = img_path.stem
    ext  = img_path.suffix.lower()

    new_stem = base
    k = 1
    while (new_stem + ext) in used_names or (out_img_dir / (new_stem + ext)).exists():
        new_stem = f"{base}__{k}"
        k += 1

    used_names.add(new_stem + ext)

    out_img = out_img_dir / (new_stem + ext)
    out_lbl = out_lbl_dir / (new_stem + ".txt")

    shutil.copy2(img_path, out_img)

    # label : copie si positif, sinon vide
    if lbl_path.exists() and lbl_path.stat().st_size > 0:
        shutil.copy2(lbl_path, out_lbl)
    else:
        out_lbl.write_text("")


def count_split(out_root: Path, split: str):
    lbl_dir = out_root / split / "labels"
    pos = 0
    bg  = 0

    for p in lbl_dir.glob("*.txt"):
        if p.stat().st_size > 0:
            pos += 1
        else:
            bg += 1

    return pos, bg, pos + bg


# =========================
# RUN
# =========================
in_root  = Path(INPUT_ROOT)
out_root = Path(OUTPUT_ROOT)

# sÃ©curitÃ©: Ã©vite Ã©crasement
if out_root.exists() and any(out_root.iterdir()):
    raise SystemExit(
        f"âš ï¸ OUTPUT_ROOT existe et n'est pas vide : {out_root}\n"
        f"Supprime-le ou change OUTPUT_ROOT."
    )

ensure_dirs(out_root)

pos_groups, bg_groups = collect_all_pairs_grouped(in_root)

pos_bases = list(pos_groups.keys())
bg_bases  = list(bg_groups.keys())

print("Bases positifs (familles):", len(pos_bases))
print("Bases backgrounds (familles):", len(bg_bases))

# 1) Split sur les BASES positives (group split -> zero leakage)
train_pos_bases, val_pos_bases, test_pos_bases = split_bases(
    pos_bases, TRAIN_P, VAL_P, TEST_P
)

# 2) Choisir des BACKGROUND bases selon ratio
def count_images_in_bases(groups, bases, allow_aug=True):
    c = 0
    for b in bases:
        for _, _, is_aug in groups[b]:
            if allow_aug or (not is_aug):
                c += 1
    return c

# RÃ¨gle NO-LEAK : VALID/TEST -> originaux seulement (no aug)
n_pos_train = count_images_in_bases(pos_groups, train_pos_bases, allow_aug=True)
n_pos_val   = count_images_in_bases(pos_groups, val_pos_bases,   allow_aug=False)
n_pos_test  = count_images_in_bases(pos_groups, test_pos_bases,  allow_aug=False)

# Backgrounds : bases distinctes (sans doublons entre splits si possible)
remaining_bg = set(bg_bases)

train_bg_bases = pick_background_bases(remaining_bg, n_pos_train * BG_RATIO_TRAIN)
remaining_bg -= train_bg_bases

val_bg_bases   = pick_background_bases(remaining_bg, n_pos_val * BG_RATIO_VAL)
remaining_bg -= val_bg_bases

test_bg_bases  = pick_background_bases(remaining_bg, n_pos_test * BG_RATIO_TEST)

# 3) Copier fichiers
used_train, used_val, used_test = set(), set(), set()

def copy_group_set(split_name, pos_base_set, bg_base_set, allow_aug_pos):
    out_img_dir = out_root / split_name / "images"
    out_lbl_dir = out_root / split_name / "labels"

    used = (
        used_train if split_name == "train"
        else used_val if split_name == "valid"
        else used_test
    )

    # POS
    for b in pos_base_set:
        for img_path, lbl_path, is_aug in pos_groups[b]:
            if allow_aug_pos or (not is_aug):
                safe_copy(img_path, lbl_path, out_img_dir, out_lbl_dir, used)

    # BG
    for b in bg_base_set:
        for img_path, lbl_path, is_aug in bg_groups[b]:
            # strict: on ignore aussi les backgrounds augmentÃ©s en valid/test
            if split_name == "train":
                safe_copy(img_path, lbl_path, out_img_dir, out_lbl_dir, used)
            else:
                if not is_aug:
                    safe_copy(img_path, lbl_path, out_img_dir, out_lbl_dir, used)

copy_group_set("train", train_pos_bases, train_bg_bases, allow_aug_pos=True)
copy_group_set("valid", val_pos_bases,   val_bg_bases,   allow_aug_pos=False)
copy_group_set("test",  test_pos_bases,  test_bg_bases,  allow_aug_pos=False)

# 4) Stats
print("\nğŸ“Š STATS FINAL NO-LEAK\n")
for sp in ["train", "valid", "test"]:
    p, b, t = count_split(out_root, sp)
    ratio = (b / p) if p else 0
    print(f"{sp.upper():5s}  pos={p}  bg={b}  total={t}  ratio bg:pos={ratio:.2f}")

print("\nâœ… Dataset NO-LEAK prÃªt :", out_root)
print("Valid/Test = originaux seulement, Train = originaux + augmentÃ©s")
```
